#### Prepping features for Psi-tomasi

```python
# params for ShiTomasi corner detection
feature_params = dict( maxCorners = 100,  # number of maximum corners
                       qualityLevel = 0.001, # minimum quality of corner below which everyone is rejected
                       minDistance = 3, # Minimum possible Euclidean distance between the returned corners
                       blockSize = 7  # Size of an average block for computing a derivative covariation matrix over each pixel neighborhood. See cornerEigenValsAndVecs .
                     )

# Parameters for lucas kanade optical flow
# docs: https://docs.opencv.org/3.4/dc/d6b/group__video__track.html#ga473e4b886d0bcc6b65831eb88ed93323
# c riteria means that at each level, the algorithm can either stop due to more number of iteratins or because error has started to drop
lk_params = dict( winSize  = (15,15), # size of the search window at each pyramid level.
                  maxLevel = 0, # 	0-based maximal pyramid level number; if set to 0, pyramids are not used (single level), if set to 1, two levels are used, and so on; if pyramids are passed to input then algorithm will use as many levels as pyramids have but no more than maxLevel.
                  criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))
```
