### Now, firstly, we need to establish a pipeline to find all the corner points quickly and gto display them

```python
path_to_folder=f"/home/anmolagarwal/Desktop/cv_assignment_2/data/all-frames-colour/{picture_type}"
```

```python
image_path_1=path_to_folder+"/frame07.png"
image_path_2=path_to_folder+"/frame08.png"
```

#### Prepping features for Psi-tomasi

```python
# params for ShiTomasi corner detection
feature_params = dict( maxCorners = 100,  # number of maximum corners
                       qualityLevel = 0.001, # minimum quality of corner below which everyone is rejected
                       minDistance = 3, # Minimum possible Euclidean distance between the returned corners
                       blockSize = 7  # Size of an average block for computing a derivative covariation matrix over each pixel neighborhood. See cornerEigenValsAndVecs .
                     )

# Parameters for lucas kanade optical flow
# docs: https://docs.opencv.org/3.4/dc/d6b/group__video__track.html#ga473e4b886d0bcc6b65831eb88ed93323
# c riteria means that at each level, the algorithm can either stop due to more number of iteratins or because error has started to drop
lk_params = dict( winSize  = (15,15), # size of the search window at each pyramid level.
                  maxLevel = 0, # 	0-based maximal pyramid level number; if set to 0, pyramids are not used (single level), if set to 1, two levels are used, and so on; if pyramids are passed to input then algorithm will use as many levels as pyramids have but no more than maxLevel.
                  criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))
```

```python
img_g_1=fetch_image(image_path_1)
img_g_2=fetch_image(image_path_2)
```

```python
# Helpful blog: https://docs.opencv.org/3.4/d4/d8c/tutorial_py_shi_tomasi.html
# Take first frame and find corners in it
p0 = cv2.goodFeaturesToTrack(img_g_1, mask = None, **feature_params)
```

```python
len(p0)
```

```python
p0
```

### Displaying the corners found via an OpenCV function

```python
corners = np.int0(p0)
plot_img=np.array(fetch_image(image_path_1, True))
for i in corners:
    x,y = i.ravel()
    cv2.circle(plot_img,(x,y),5,(255,0,0),-1)
plt.rcParams["figure.figsize"] = (10,10)

plt.imshow(plot_img)
```

```python
print(corners)
```

### LK now

```python
# calculate optical flow
#  It returns next points along with some status numbers which has a value of 1 if next point is found, else zero.

# st: output status vector (of unsigned chars); each element of the vector is set to 1 if the flow for the corresponding features has been found, otherwise, it is set to 0.
p1, st, err = cv2.calcOpticalFlowPyrLK(img_g_1,
                                       img_g_2,
                                       p0,
                                       None,
                                       **lk_params)
```

```python
len(p1)
```

```python
st
```

```python
img_g_1.shape
```

```python
# Select good points
good_new = p1[st==1]
good_old = p0[st==1]
```

```python
# draw the tracks
color = np.random.randint(0,255,(10000,3))
mask = np.zeros_like(fetch_image(image_path_1, True))
frame=fetch_image(image_path_1, True)
u=np.zeros(fetch_image(image_path_1).shape)
v=np.zeros(fetch_image(image_path_1).shape)
for i,(old,new) in enumerate(zip(good_old,good_new)):
    a,b = old.ravel()
    c,d = new.ravel()
    a=int(a)
    b=int(b)
    c=int(c)
    d=int(d)
    
    a,b=b,a
    c,d=d,c
    try:
        print("#########")
        print(f"changing from {u[a][b]} to {c-a}")
        u[a][b]=c-a
        v[a][b]=d-b
        mask = cv2.line(mask, (b,a),(d,c), color[i].tolist(), 2)
        frame = cv2.circle(frame,(b,a),5,color[i].tolist(),-1)
        print(f"{a}:{b}:{c}:{d} --> {c-a} :: {d-b}")
        #print("u max is ", u.max())
        print("#########")
    except Exception as E:
        pass
    #break
    
img = cv2.add(frame,mask)

plt.imshow(img)
```
